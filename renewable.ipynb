```python
import pandas as pd
```

```python
file_paths = [
    "data/IRR/ELECSTAT_20250215-174354.xlsx",
    "data/IRR/ELECSTAT_20250215-174508.xlsx",
    "data/IRR/ELECSTAT_20250215-174559.xlsx",
    "data/IRR/ELECSTAT_20250215-174615.xlsx"
]

# Read the first sheet of each Excel file to get an overview
dataframes = [pd.ExcelFile(file).parse(sheet_name=0) for file in file_paths]

# merge all dataframes
elecstats_df = pd.concat(dataframes)
elecstats_df.to_csv("data/IRR/ELECSTAT.csv", index=False)
# df.head()
elecstats_df.sample(5)
```

```python
file_paths = [
    "data/IRR/PUBFIN_20250215-174006.xlsx",
    "data/IRR/PUBFIN_20250215-174057.xlsx"
]

# Read the first sheet of each Excel file to get an overview
dataframes = [pd.ExcelFile(file).parse(sheet_name=0) for file in file_paths]

# merge all dataframes
pubfin_df = pd.concat(dataframes)
pubfin_df.to_csv("data/IRR/PUBFIN.csv", index=False)
# df.head()
pubfin_df.sample(5)
```

```python
elecstats_df_cleaned = elecstats_df.dropna(thresh=3)
elecstats_df_cleaned.ffill(axis=0, inplace=True)

elecstats_df_cleaned.columns = [
    "Region_Tech_Desc",
    "Category",
    "Data_Type",
    "Year",
    "Electricity_Output_GWh",
    "Grid_Connection",
    "Miscellaneous",
]
elecstats_df_cleaned.reset_index(drop=True, inplace=True)


elecstats_df_cleaned.to_csv("data/IRR_cleaned/ELECSTAT_CLEANED.csv", index=False)
elecstats_df_cleaned.sample(5)
```

```python
heat_gen_df = pd.read_excel("data/IRR/HEATGEN_20250215-174254.xlsx", skiprows=1, header=None)

heat_gen_df.columns = [
    "Country/Area",
    "Technology",
    "Grid Connection",
    "Year",
    "Heat Generation (TJ)",
]


heat_gen_df["Country/Area"].fillna(method="ffill", inplace=True)
heat_gen_df["Technology"].fillna(method="ffill", inplace=True)
heat_gen_df['Grid Connection'].fillna(method="ffill", inplace=True)


heat_gen_df_cleaned = heat_gen_df.dropna(subset=["Year", "Heat Generation (TJ)"])


heat_gen_df_cleaned["Year"] = heat_gen_df_cleaned["Year"].astype(int)
heat_gen_df_cleaned["Heat Generation (TJ)"] = pd.to_numeric(
    heat_gen_df_cleaned["Heat Generation (TJ)"], errors="coerce"
)
# TJ stands for terajoules, which is a unit of energy

heat_gen_df_cleaned = heat_gen_df_cleaned.dropna(subset=["Heat Generation (TJ)"])


heat_gen_df_cleaned.reset_index(drop=True, inplace=True)

heat_gen_df_cleaned.to_csv("data/IRR_cleaned/HEATGEN_CLEANED.csv", index=False)
heat_gen_df_cleaned.sample(5)
```

```python
reshare_df = pd.read_excel("data/IRR/RESHARE_20250215-173602.xlsx", sheet_name="RESHARE")

reshare_df_cleaned = reshare_df.dropna(how="all")


reshare_df_cleaned.iloc[:, 0] = reshare_df_cleaned.iloc[:, 0].fillna(method="ffill")


reshare_df_cleaned.columns = ["Region/Country", "Indicator", "Year", "Value"]


reshare_df_cleaned = reshare_df_cleaned.dropna(subset=["Year", "Value"])


reshare_df_cleaned["Year"] = reshare_df_cleaned["Year"].astype(int)


reshare_df_cleaned["Value"] = pd.to_numeric(reshare_df_cleaned["Value"], errors="coerce")


reshare_df_cleaned = reshare_df_cleaned.dropna(subset=["Value"])

reshare_df_cleaned.reset_index(drop=True, inplace=True)
reshare_df_cleaned.to_csv("data/IRR_cleaned/RESHARE_CLEANED.csv", index=False)
reshare_df_cleaned.sample(5)
```

```python
pubfin_df.columns = ["Country", "Technology", "Year", "Investment"]


pubfin_df["Country"] = pubfin_df["Country"].fillna(method="ffill")
pubfin_df["Technology"] = pubfin_df["Technology"].fillna(method="ffill")


pubfin_df_cleaned = pubfin_df.dropna(subset=["Year", "Investment"])


pubfin_df_cleaned["Year"] = pubfin_df_cleaned["Year"].astype(int)
pubfin_df_cleaned["Investment"] = pubfin_df_cleaned["Investment"].astype(float)

pubfin_df_cleaned.reset_index(drop=True, inplace=True)
pubfin_df_cleaned.to_csv("data/IRR_cleaned/PUBFIN_CLEANED.csv", index=False)
pubfin_df_cleaned.sample(5)
```
